```python
import pandas as pd
```

```python
df=pd.read_csv('/kaggle/input/data-nlp/yelp.csv')
```

```python
df.head()
```

```python
df['text'][0]
```

```python
import seaborn as sns
```

```python
sns.countplot(x=df['stars'])
```

```python
# convert tolowercase,the text and replace unnecesry punctuations.
```

```python
df['text']=df['text'].str.lower()
df['text']=df['text'].str.replace('[^\w\s]','') #[^\w\s] finds all punctiatuins.
df['text']=df['text'].str.replace('\d+','') #\d+ removes numbers
df['text']=df['text'].str.replace('\n','').replace('\r','') # newlines \n is removed,\r(enter)is also removed.
```

```python
df.head()
```

```python
df2=df[(df['stars']==1)|(df['stars']==5)]
df2.head()
```

```python
# we are gonna clean the bad reviews.
```

```python
x=df2[['text']]
y=df2['stars']
```

```python
x
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
vect=CountVectorizer(stop_words='english',ngram_range=(1,2),min_df=2) #ngram for ccontext ,(1,2)means 2 words together ittskes.
x_v=vect.fit_transform(x.text)
tf=pd.DataFrame(x_v.toarray(),columns=vect.get_feature_names_out())
tf
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report ,accuracy_score
x_train,x_test,y_train,y_test=train_test_split(x_v,y,random_state=42,test_size=0.20)
```

```python
from sklearn.naive_bayes import MultinomialNB # data not 1 or 0 1-or 5
nb=MultinomialNB()
nb.fit(x_train,y_train)
prednb=nb.predict(x_test)
print(classification_report(prednb,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
predlr=lr.predict(x_test)
print(classification_report(predlr,y_test))
```
