# modelling

```python
df.head()
```

```python
x=df.iloc[:,:2]
y=df.iloc[:,2:]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.20)

dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
pred=dtc.predict(x_test)
print(classification_report(pred,y_test))
```
