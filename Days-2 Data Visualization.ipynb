```python
x=np.arange(20)
y=np.random.normal(10,1,20) # 10=>mean,1=>S.D,20=>no of numbers.#generate 20numbers
z=np.random.normal(10,2,20)
k=np.random.normal(10,3,20)
```

```python
plt.plot(x,y,color='green',label='y'); # ';' is used to hide location.
plt.plot(x,z,color='y',label='z');
plt.plot(x,k,color='r',label='k');
plt.legend(loc='upper left');
# legend isthe box that displays the color and label of the plots.
#default location ofthe legend is upper right,we can use loc(location) to specify the location as well. 
```

```python
plt.show(plt.plot(x,y)) #plt.show()
```

```python
##making separte plots into charts
fig=plt.figure(figsize=(20,5)) #(10,5)=> plotsize in inches.(width,height)
ax=fig.add_subplot(131) # 1=row,3=coloumns,1=first chart
ax.set_title('X vs Y') #for labeling
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.plot(x,y,)

bx=fig.add_subplot(132) 
bx.set_title('X vs Z')
bx.set_xlabel('X values')
bx.set_ylabel('Z values')
bx.plot(x,z)

cx=fig.add_subplot(133) 
cx.set_title('X vs K')
cx.set_xlabel('X values')
cx.set_ylabel('K values')
cx.plot(x,k);
```

```python
##making separte plots into charts
fig=plt.figure(figsize=(20,10)) #(10,5)=> plotsize in inches.(width,height)
ax=fig.add_subplot(231) # 1=row,3=coloumns,1=first chart
ax.set_title('X vs Y') #for labeling
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.plot(x,y)

bx=fig.add_subplot(232) 
bx.set_title('X vs Z')
bx.set_xlabel('X values')
bx.set_ylabel('Z values')
bx.plot(x,z)

cx=fig.add_subplot(233) 
cx.set_title('X vs K')
cx.set_xlabel('X values')
cx.set_ylabel('K values')
cx.scatter(z,k);

dx=fig.add_subplot(234) 
dx.set_title('X vs K')
dx.set_xlabel('X values')
dx.set_ylabel('K values')
dx.plot(x,k);

ex=fig.add_subplot(235) 
ex.set_title('X vs K')
ex.set_xlabel('X values')
ex.set_ylabel('K values')
ex.plot(x,k);

fx=fig.add_subplot(236) 
fx.set_title('X vs K')
fx.set_xlabel('X values')
fx.set_ylabel('K values')
fx.boxplot([y,z,k]);

```

# <font color=red><b>Feature engineering</b>
    

```python
df['day']=df['Date'].dt.day #created new column nmaed day from the date dataset.this is feature engg.
```

```python
df.head()
```

```python
df['month']=df['Date'].dt.month
```

```python
df.head()
```

```python
df['year']=df['Date'].dt.year
df['month_name']=df['Date'].dt.month_name()
df['weekday']=df['Date'].dt.day_name()
df.head()
```

```python
sns.countplot(x=df['Gender']);
```

```python
plt.figure(figsize=(8,5));
sns.countplot(x=df['weekday']);
plt.xticks(rotation=90);
```
