```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/kaggle/input/loan-apporval-prediction/loan_prediction.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
df['Married'].fillna(df['Married'].mode()[0],inplace=True)
df['Dependents'].fillna(df['Dependents'].mode()[0],inplace=True)
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0],inplace=True)
df['LoanAmount'].fillna(df['LoanAmount'].mode()[0],inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0],inplace=True)
df['Credit_History'].fillna(df['Credit_History'].mode()[0],inplace=True)
```

```python
df=df.drop('Loan_ID',axis=1)
```

```python
dc={'Y':1,'N':0}
df['Loan_Status']=df['Loan_Status'].map(dc)
```

```python
df.head()
```

```python
x=df.drop('Loan_Status',axis=1)
y=df['Loan_Status']
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x.info()
```

```python
from sklearn.metrics import accuracy_score
```

```python

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB,BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

g,b,l,d,r,svc,knn=GaussianNB(),BernoulliNB(),LogisticRegression(),DecisionTreeClassifier(),RandomForestClassifier(),SVC(),KNeighborsClassifier()
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.20)

```

```python
g.fit(x_train,y_train)
b.fit(x_train,y_train)
l.fit(x_train,y_train)
d.fit(x_train,y_train)
r.fit(x_train,y_train)
svc.fit(x_train,y_train)
knn.fit(x_train,y_train)




predg=g.predict(x_test)
predb=b.predict(x_test)
predl=l.predict(x_test)
predd=d.predict(x_test)
predr=r.predict(x_test)
predsvc=svc.predict(x_test)
predknn=knn.predict(x_test)
```

```python
accuracy_score(y_test,predg),accuracy_score(y_test,predb),accuracy_score(y_test,predl),accuracy_score(y_test,predd),accuracy_score(y_test,predr),accuracy_score(y_test,predsvc),accuracy_score(y_test,predknn)
```
